```python
# read the data into pandas dataframes
CV_T00_GF_106a = pd.read_csv("lca/cleaned/106_CV_T00_GF_lca.csv")

# delete any Metazoa hits because they're trypsin
CV_T00_GF_106b = CV_T00_GF_106a[CV_T00_GF_106a.kingdom != 'Metazoa']

# keep only entries to the phylum level
CV_T00_GF_106 = CV_T00_GF_106b[CV_T00_GF_106b['phylum'].notnull()]

# How many peptides to the phylum level?
print('# of phylum peptides = ', len(CV_T00_GF_106))

CV_T00_GF_106.head(6)
```

### Now I want to read in the file containing the stripped peptides with NAAF values
####  - NAAF stands for 'noramlized area abunace factor'

### I want to join the dataframes if they share an index (stripped peptide with equated leucine and isoleucines)
#### - That means I'll reindex the processed peptide file

```python
CV_T00_GF_106_NAAFa = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/I-L_NAAFs/106A_CV_T00_GF_DN50_ILnaafs.csv")
CV_T00_GF_106_NAAFb = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/I-L_NAAFs/106B_CV_T00_GF_DN50_ILnaafs.csv")
CV_T00_GF_106_NAAFc = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/I-L_NAAFs/106C_CV_T00_GF_DN50_ILnaafs.csv")


frames = [CV_T00_GF_106_NAAFa, CV_T00_GF_106_NAAFb, CV_T00_GF_106_NAAFc]

CV_T00_GF_106_NAAF = pd.concat(frames, sort=False)

CV_T00_GF_106_NAAF.set_index('stripped_peptide')
CV_T00_GF_106_NAAF = CV_T00_GF_106_NAAF.loc[:, ~CV_T00_GF_106_NAAF.columns.str.contains('^Unnamed')]

CV_T00_GF_106_NAAF.rename(columns = {'stripped_peptide':'peptide'}, inplace = True)

print('# of total peptides = ', len(CV_T00_GF_106_NAAF))

print('column names:', CV_T00_GF_106_NAAF.columns)

CV_T00_GF_106_NAAF.head()
```

```python
# get the intersection of the phylum peptides and corresponding peptides w/ NAAFs

over = np.intersect1d(CV_T00_GF_106["peptide"], CV_T00_GF_106_NAAF["peptide"])

print(over)
```

```python
# join the dataframes if the peptide values are the same using 'join'
# since a couple are de novo only (more for bacteria), we won't have all the UniPept peptides overlap 

CV_T00_GF_106.set_index('peptide', inplace=True)
CV_T00_GF_106_NAAF.set_index('peptide', inplace=True)

CV_T00_GF_106_Phy = CV_T00_GF_106.join(CV_T00_GF_106_NAAF, how='left', rsuffix='_other')


print('# of total phylum-level peptides = ', len(CV_T00_GF_106_Phy))

CV_T00_GF_106_Phy.head()
```

```python
# write to a csv

CV_T00_GF_106_Phy.to_csv("lca/NAAF/CV_T00_GF_106_Phy_naaf.csv")
```
