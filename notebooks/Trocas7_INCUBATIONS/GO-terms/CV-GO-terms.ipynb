## Evaluating and visualizing gene ontology (GO) terms associated bacterial peptides in Trocas 7 incubations

#### Beginning with: stripped bacterial peptides and GO term abundances in spreadsheets

#### Goal: heatmaps of GO terms over stations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from functools import reduce
```

## Plotting bacterial GO terms using Seaborn heatmap

### For each incubation station
### For all 3 classes of GO: biological process, cellular component, molecular function

```python
# rearranged and condensed the csv manually in Libre Calc
# read in new csv:

cc = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/analyses/T7-incubations/unipept/GO/plotting/CV_CC_GO_sorted.csv", index_col=0)

cc.head()
```

```python
sns.set(font_scale=1.2)  

fig, ax = plt.subplots(figsize=(4,7))

labels = ['Time 0 small', 'Time 0 large', '', 'Time 24 small', 'Time 24 large']

ax = sns.heatmap(cc, cmap="BuPu", annot=True, linecolor = 'grey', linewidths=.01, \
                square=False, cbar_kws={'label': '% of GO terms in sample'}, label='big')

ax.set_ylabel('')

# turn the axis label
for item in ax.get_yticklabels():
    item.set_rotation(0)
    

for item in ax.get_xticklabels():
    item.set_rotation(45)
    
ax.set_xticklabels(labels, rotation = 50)
    
plt.title("% of cellular component GO terms: Chaves", size = 16, pad = 20)
```
