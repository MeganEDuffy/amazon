## Plotting microbial GO terms using Seaborn heatmaps

### Just molecuar function GO terms

```python
# rearranged and condensed the csv manually in Libre Calc
# read in new csv:

MN = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/analyses/T7-incubations/unipept/GO/plotting/MN_MF_GO_sorted.csv", index_col=0)
MS = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/analyses/T7-incubations/unipept/GO/plotting/MS_MF_GO_sorted.csv", index_col=0)
BY = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/analyses/T7-incubations/unipept/GO/plotting/BY_MF_GO_sorted.csv", index_col=0)
CV = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/analyses/T7-incubations/unipept/GO/plotting/CV_MF_GO_sorted.csv", index_col=0)

CV.head()
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,sharey=True, sharex = True, figsize = (9, 5))

from matplotlib import style

style.use("seaborn-darkgrid")

plt.tight_layout()
plt.subplots_adjust(wspace = 0.13)


labels = ['0 hr', '24 hr', '', '0 hr', '24 hr']

cbar_ax = fig.add_axes([1.0, 0.04,.03, 0.92])
cbar_ax.set_xlabel('% of all sample GO terms', labelpad = 20)

g1 = sns.heatmap(MN,cmap="Blues", cbar_ax = cbar_ax ,ax=ax1, square = True, annot = True)
g1.set_xlabel('0.3 - 0.7 $\mu$m      >0.7 $\mu$m', labelpad = 15)
g1.set_title('Macapá North', fontsize = 16, loc='center', pad = 10)
g1.set_xticklabels(labels, rotation = 45, fontsize = 10)


g2 = sns.heatmap(MS,cmap="Blues", cbar_ax = cbar_ax, ax=ax2, square = True,annot = True)
g2.set_ylabel('')
g2.set_xlabel('0.3 - 0.7 $\mu$m      >0.7 $\mu$m', labelpad = 15)
g2.set_title('Macapá South', fontsize = 16, loc='center', pad = 10)
g2.set_xticklabels(labels, rotation = 45, fontsize = 10)


g3 = sns.heatmap(BY,cmap="Blues", cbar_ax = cbar_ax, ax=ax3, square = True, annot = True)
g3.set_ylabel('')
g3.set_xlabel('0.3 - 0.7 $\mu$m      >0.7 $\mu$m', labelpad = 15)
g3.set_title('Baylique', fontsize = 16, loc='center', pad = 10)
g3.set_xticklabels(labels, rotation = 45, fontsize = 10)


g4 = sns.heatmap(CV,cmap="Blues",ax=ax4, cbar_ax = cbar_ax, edgecolor = 'black', square = True, annot = True)
g4.set_ylabel('')
g4.set_xlabel('0.3 - 0.7 $\mu$m      >0.7 $\mu$m', labelpad = 15)
g4.set_title('Chaves', fontsize = 16, loc='center', pad = 10)
g4.set_xticklabels(labels, rotation = 45, fontsize = 10)

# add labels to y axis 

g1.set_ylabel('Organic comp. binding - Metal binding - Energy prod. - Growth', labelpad = 10, fontsize = 13);
```

```python
# for thesis talk

fig,(ax1) = plt.subplots(1,1, figsize = (7, 7))

from matplotlib import style

style.use("seaborn-darkgrid")

plt.tight_layout()
plt.subplots_adjust(wspace = 0.13)

labels = ['0 hr', '24 hr', '', '0 hr', '24 hr']
ylabels = ['ATP binding', 'ATPase', 'DNA binding', 'RNA binding', 'catalytic activity', \
          'hydrolase activity', 'oxidoreductase activity', 'metal ion binding', 'zinc ion binding', \
          'calcium ion binding', 'iron ion binding', 'transmembrane transporter', 'peptidase activity', \
          'carbohydrate binding', 'chitin binding', 'other']

cbar_ax = fig.add_axes([1.09, 0.05,.07, 0.92])
cbar_ax.set_xlabel('% of all peptides', labelpad = 20, fontsize = 16)
cbar_ax.tick_params(labelsize=16)

g1 = sns.heatmap(MN,cmap="Blues", cbar_ax = cbar_ax ,ax=ax1, square = False, annot = True, annot_kws={"size": 14})
g1.set_xlabel('0.3 - 0.7 $\mu$m                                    >0.7 $\mu$m', labelpad = 15, fontsize = 16)
g1.set_title('Chaves (South Channel Mouth)', fontsize = 20, loc='center', pad = 10)
g1.set_xticklabels(labels, rotation = 45, fontsize = 16)
g1.set_yticklabels(ylabels, fontsize = 16)


# add labels to y axis 

g1.set_ylabel('Organic comp. binding - Metal binding - Energy prod. - Growth', labelpad = 10, fontsize = 16);

plt.savefig('/home/millieginty/Documents/git-repos/2017-etnp/data/thesis-talk/chaves-mf.png', bbox_inches='tight', dpi=400)
```
