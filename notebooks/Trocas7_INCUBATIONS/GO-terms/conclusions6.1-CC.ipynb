### Goal: stacked bar plots of GO terms over all these samples

### Data location: git-repos/general-exam/dissertation/meta-analysis/meta-xchapter-cellular-go-fig6.xlsx

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from functools import reduce
```

```python
#import cross-chapter combined CC data

XC = pd.read_csv("/home/millieginty/Documents/git-repos/general-exam/dissertation/meta-analysis/cross-chapter-cc.csv", index_col=0)

XC_T = XC.transpose()

XC_T.head(20)
```

```python
# plot stacked bar

fig, ax = plt.subplots(figsize=(10, 5))

colors = ['#2ca25f', '#bdbdbd', '#2b8cbe', '#de2d26', '#bcbddc', '#bf812d', '#c7eae5', '#f6e8c3']

plt.tight_layout()

labels = ['Prochloroccus culture', 'Diatom rot Day 0', 'Diatom rot Day 12', 'ETNP 100 m sus POM', \
          'ETNP 100 m sink POM', 'ETNP 965 m sus POM',\
          'ETNP 965 m sink POM', 'ETNP sediment core', \
          'Amazon River POM', '', 'ETNP 100 m sus POM algal', 'Deep ETNP algal', 'ETNP sediment algal',\
         'Amazon River algal']

#  Core total
XC_T.plot(ax = ax, kind='bar', stacked=True, color = colors, edgecolor = 'black')
#ax.set_title('Culture and environmental peptide cellular compartment GO terms', fontsize = 20, loc='left')
ax.set_ylabel("% cellular component GO terms", labelpad=10, fontsize = 16)
ax.set_xticklabels(labels, rotation = 90, fontsize = 16)
ax.get_legend().remove()
ax.tick_params(axis='y', labelsize= 14)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::1], labels[::1], bbox_to_anchor=(1,1), \
                 loc="upper left", fontsize = 16, ncol = 1)
```
