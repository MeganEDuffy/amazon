### Manipulation of PeaskDB de novo-assisted database search results of Trocas 7 (high water, April 2019) incubation samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the TROCAS # (e.g., 101) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup110 = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/T7-INC-UWPR_Apr2021_Fus_T00T24/DN50_10lgP15/110_TROCAS7_Fusion_Apr2021_PEAKS_146/110_TROCAS7_Fusion_Apr2021_PEAKS_146_peptide.csv")

# remove redundant rows
peaksdb110 = pd.DataFrame.drop_duplicates(peaksdbdup110)

print(peaksdb110.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source_File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb110.columns = columns

#remmove # spec and accession columns because they mess parsing up

del peaksdb110['#Spec']
del peaksdb110['Accession']
del peaksdb110['PTM']
del peaksdb110['AScore']

mean_length = peaksdb110['Length'].mean()
print(mean_length)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup110))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb110))

#look at the dataframe
peaksdb110.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb110['A'] = peaksdb110['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb110['C'] = peaksdb110['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb110['D'] = peaksdb110['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb110['E'] = peaksdb110['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb110['F'] = peaksdb110['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb110['G'] = peaksdb110['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb110['H'] = peaksdb110['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb110 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb110['I'] = peaksdb110['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb110['K'] = peaksdb110['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb110['L'] = peaksdb110['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb110['M'] = peaksdb110['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb110['N'] = peaksdb110['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb110['P'] = peaksdb110['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb110['Q'] = peaksdb110['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb110['R'] = peaksdb110['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb110['S'] = peaksdb110['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb110['T'] = peaksdb110['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb110['V'] = peaksdb110['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb110['W'] = peaksdb110['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb110['Y'] = peaksdb110['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb110['c-carb'] = peaksdb110['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb110['m-oxid'] = peaksdb110['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb110['n-deam'] = peaksdb110['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a lamba function to enumerate the # of deamidated Q's in each peptide
peaksdb110['q-deam'] = peaksdb110['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb110['stripped_peptide'] = peaksdb110['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb110['stripped_length'] = peaksdb110['stripped_peptide'].apply(len)

peaksdb110['NAAF_num.'] = peaksdb110['Area'] / peaksdb110['stripped_length']

# total the number of modifications in sequence
peaksdb110['ptm-total'] = peaksdb110['c-carb'] + peaksdb110['m-oxid'] + peaksdb110['n-deam'] + peaksdb110['q-deam']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb110['stripped_IL']= peaksdb110['stripped_peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb110.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/PeaksDB/110_MS_T00_GF_PDB.csv")

# check out the results
peaksdb110.head()
```

```python
# keep only stripped peptide I/L and NAAF
db_110 = peaksdb110[['stripped_IL', 'Area', 'NAAF_num.']]

db_110.set_index('stripped_IL')

# write modified dataframe to new txt file
db_110.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/I-L_NAAFs/110_MS_T00_GF_PDB_ILnaafs.csv")

db_110.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb110['A'].sum(),
        'C': peaksdb110['C'].sum(),
        'D': peaksdb110['D'].sum(),
        'E': peaksdb110['E'].sum(),
        'F': peaksdb110['F'].sum(),
        'G': peaksdb110['G'].sum(),
        'H': peaksdb110['H'].sum(),
        'I': peaksdb110['I'].sum(),
        'K': peaksdb110['K'].sum(),
        'L': peaksdb110['L'].sum(),
        'M': peaksdb110['M'].sum(),
        'N': peaksdb110['N'].sum(),
        'P': peaksdb110['P'].sum(),
        'Q': peaksdb110['Q'].sum(),
        'R': peaksdb110['R'].sum(),
        'S': peaksdb110['S'].sum(),
        'T': peaksdb110['T'].sum(),
        'V': peaksdb110['V'].sum(),
        'W': peaksdb110['W'].sum(),
        'Y': peaksdb110['Y'].sum(),
        'c-carb': peaksdb110['c-carb'].sum(),
        'm-oxid': peaksdb110['m-oxid'].sum(),
        'n-deam': peaksdb110['n-deam'].sum(),
        'q-deam': peaksdb110['n-deam'].sum(),
        'Total area': peaksdb110['Area'].sum(),
        'Total length': peaksdb110['stripped_length'].sum()
       }

totalpeaksdb110 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 
                                              'n-deam', 'q-deam', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb110['% C w/ carb'] = totalpeaksdb110['c-carb'] / totalpeaksdb110['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb110['% M w/ oxid'] = totalpeaksdb110['m-oxid'] / totalpeaksdb110['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb110['% N w/ deam'] = totalpeaksdb110['n-deam'] / totalpeaksdb110['N'] 

# calculate percentage of N's that are deamidated
totalpeaksdb110['% Q w/ deam'] = totalpeaksdb110['n-deam'] / totalpeaksdb110['N'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb110['NAAF denom.'] = totalpeaksdb110['Total area'] / totalpeaksdb110['Total length']

# write modified dataframe to new txt file
totalpeaksdb110.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/PeaksDB/110_MS_T00_GF_PDB_totals.csv")

totalpeaksdb110.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb110 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 209651.289817

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb110['NAAF factor'] = (peaksdb110['NAAF_num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb110_NAAF = peaksdb110[['stripped_peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'n-deam', 'q-deam']].copy()

# multiply the NAAF20 factor by the AAs to normalize its abundance by peak area and peptide length

peaksdb110_NAAF['A-NAAF20'] = peaksdb110_NAAF['A'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['C-NAAF20'] = peaksdb110_NAAF['C'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['D-NAAF20'] = peaksdb110_NAAF['D'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['E-NAAF20'] = peaksdb110_NAAF['E'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['F-NAAF20'] = peaksdb110_NAAF['F'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['G-NAAF20'] = peaksdb110_NAAF['G'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['H-NAAF20'] = peaksdb110_NAAF['H'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['I-NAAF20'] = peaksdb110_NAAF['I'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['K-NAAF20'] = peaksdb110_NAAF['K'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['L-NAAF20'] = peaksdb110_NAAF['L'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['M-NAAF20'] = peaksdb110_NAAF['M'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['N-NAAF20'] = peaksdb110_NAAF['N'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['P-NAAF20'] = peaksdb110_NAAF['P'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['Q-NAAF20'] = peaksdb110_NAAF['Q'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['R-NAAF20'] = peaksdb110_NAAF['R'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['S-NAAF20'] = peaksdb110_NAAF['S'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['T-NAAF20'] = peaksdb110_NAAF['T'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['V-NAAF20'] = peaksdb110_NAAF['V'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['W-NAAF20'] = peaksdb110_NAAF['W'] * peaksdb110['NAAF factor']
peaksdb110_NAAF['Y-NAAF20'] = peaksdb110_NAAF['Y'] * peaksdb110['NAAF factor']

# multiply the NAAF20 factor by the PTMs normalize its abundance by peak area and peptide length

peaksdb110_NAAF['ccarb-NAAF20'] = peaksdb110_NAAF['c-carb'] * peaksdb110_NAAF['NAAF factor']
peaksdb110_NAAF['moxid-NAAF20'] = peaksdb110_NAAF['m-oxid'] * peaksdb110_NAAF['NAAF factor']
peaksdb110_NAAF['ndeam-NAAF20'] = peaksdb110_NAAF['n-deam'] * peaksdb110_NAAF['NAAF factor']
peaksdb110_NAAF['qdeam-NAAF20'] = peaksdb110_NAAF['q-deam'] * peaksdb110_NAAF['NAAF factor']


# write the dataframe to a new csv
peaksdb110_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/PeaksDB/110_MS_T00_GF_PDB_naaf.csv")

peaksdb110_NAAF.head()
```

```python
# made a new dataframe that contains the sums of NAAF normalized AAs for peaksdb110 peaksdb results
# also contains the sums of the NAAF-corrected PTMs occurances for each affected residue

index = ['sample total']

data = {'NAAF': peaksdb110_NAAF['NAAF factor'].sum(),
        'A': peaksdb110_NAAF['A-NAAF20'].sum(),
        'C': peaksdb110_NAAF['C-NAAF20'].sum(),
        'D': peaksdb110_NAAF['D-NAAF20'].sum(),
        'E': peaksdb110_NAAF['E-NAAF20'].sum(),
        'F': peaksdb110_NAAF['F-NAAF20'].sum(),
        'G': peaksdb110_NAAF['G-NAAF20'].sum(),
        'H': peaksdb110_NAAF['H-NAAF20'].sum(),
        'I': peaksdb110_NAAF['I-NAAF20'].sum(),
        'K': peaksdb110_NAAF['K-NAAF20'].sum(),
        'L': peaksdb110_NAAF['L-NAAF20'].sum(),
        'M': peaksdb110_NAAF['M-NAAF20'].sum(),
        'N': peaksdb110_NAAF['N-NAAF20'].sum(),
        'P': peaksdb110_NAAF['P-NAAF20'].sum(),
        'Q': peaksdb110_NAAF['Q-NAAF20'].sum(),
        'R': peaksdb110_NAAF['R-NAAF20'].sum(),
        'S': peaksdb110_NAAF['S-NAAF20'].sum(),
        'T': peaksdb110_NAAF['T-NAAF20'].sum(),
        'V': peaksdb110_NAAF['V-NAAF20'].sum(),
        'W': peaksdb110_NAAF['W-NAAF20'].sum(),
        'Y': peaksdb110_NAAF['Y-NAAF20'].sum(),
        'c-carb': peaksdb110_NAAF['ccarb-NAAF20'].sum(),
        'm-oxid': peaksdb110_NAAF['moxid-NAAF20'].sum(),
        'n-deam': peaksdb110_NAAF['ndeam-NAAF20'].sum(),
        'q-deam': peaksdb110_NAAF['qdeam-NAAF20'].sum()
       }

totalpeaksdb110_NAAF = pd.DataFrame(data, columns=['NAAF', 'A', 'C', 'D', 'E', 'F', \
                                                   'G', 'H', 'I','K', 'L', 'M', \
                                                   'N', 'P', 'Q', 'R', 'S', \
                                                   'T', 'V', 'W', 'Y', 'c-carb', \
                                                   'm-oxid', 'n-deam', 'q-deam'], index=index)

# calculate the NAAF-corrected % modified C, M, N, Q, K, and Rs


# calculate NAAF-corrected percentage of C's with carb (should be 1.0)
totalpeaksdb110_NAAF['% C w/ carb.'] = totalpeaksdb110_NAAF['c-carb'] / totalpeaksdb110_NAAF['C'] 

# calculate NAAF-corrected percentage of M's that are oxidized
totalpeaksdb110_NAAF['% M w/ oxid'] = totalpeaksdb110_NAAF['m-oxid'] / totalpeaksdb110_NAAF['M'] 

# calculate NAAF-corrected percentage of N's that are deamidated
totalpeaksdb110_NAAF['% N w/ deam'] = totalpeaksdb110_NAAF['n-deam'] / totalpeaksdb110_NAAF['N'] 

# calculate NAAF-corrected percentage of N's that are deamidated
totalpeaksdb110_NAAF['% Q w/ deam'] = totalpeaksdb110_NAAF['n-deam'] / totalpeaksdb110_NAAF['N'] 

# calculate NAAF summed numerator over denominator (in above cell) for all peptides in dataset i: a check
totalpeaksdb110_NAAF['NAAF check'] = totalpeaksdb110_NAAF['NAAF'] / 209651.289817

# write modified dataframe to new txt file, same name + totals
totalpeaksdb110_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/PeaksDB/110_MS_T00_GF_PDB_naaf_totals.csv")

totalpeaksdb110_NAAF.head()
```
