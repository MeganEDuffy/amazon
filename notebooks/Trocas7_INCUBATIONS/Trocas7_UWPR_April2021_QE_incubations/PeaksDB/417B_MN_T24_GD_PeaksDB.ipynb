### Manipulation of PeaskDB de novo-assisted database search results of Trocas 7 (high water, April 2019) incubation samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the TROCAS # (e.g., 101) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup417 = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/T7-INC-UWPR_Apr2021_Fus_T00T24/DN50_10lgP15/417_TROCAS7_Fusion_Apr2021_PEAKS_180/417_TROCAS7_Fusion_Apr2021_PEAKS_180_peptide.csv")

# remove redundant rows
peaksdb417 = pd.DataFrame.drop_duplicates(peaksdbdup417)

print(peaksdb417.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source_File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb417.columns = columns

#remmove # spec and accession columns because they mess parsing up

del peaksdb417['#Spec']
del peaksdb417['Accession']
del peaksdb417['PTM']
del peaksdb417['AScore']

mean_length = peaksdb417['Length'].mean()
print(mean_length)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup417))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb417))

#look at the dataframe
peaksdb417.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb417['A'] = peaksdb417['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb417['C'] = peaksdb417['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb417['D'] = peaksdb417['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb417['E'] = peaksdb417['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb417['F'] = peaksdb417['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb417['G'] = peaksdb417['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb417['H'] = peaksdb417['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb417 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb417['I'] = peaksdb417['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb417['K'] = peaksdb417['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb417['L'] = peaksdb417['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb417['M'] = peaksdb417['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb417['N'] = peaksdb417['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb417['P'] = peaksdb417['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb417['Q'] = peaksdb417['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb417['R'] = peaksdb417['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb417['S'] = peaksdb417['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb417['T'] = peaksdb417['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb417['V'] = peaksdb417['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb417['W'] = peaksdb417['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb417['Y'] = peaksdb417['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb417['c-carb'] = peaksdb417['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb417['m-oxid'] = peaksdb417['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb417['n-deam'] = peaksdb417['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a lamba function to enumerate the # of deamidated Q's in each peptide
peaksdb417['q-deam'] = peaksdb417['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb417['stripped_peptide'] = peaksdb417['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb417['stripped_length'] = peaksdb417['stripped_peptide'].apply(len)

peaksdb417['NAAF_num.'] = peaksdb417['Area'] / peaksdb417['stripped_length']

# total the number of modifications in sequence
peaksdb417['ptm-total'] = peaksdb417['c-carb'] + peaksdb417['m-oxid'] + peaksdb417['n-deam'] + peaksdb417['q-deam']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb417['stripped_IL']= peaksdb417['stripped_peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb417.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/PeaksDB/417_MN_T24_GD_PDB/417B_MN_T24_GD_PDB.csv")

# check out the results
peaksdb417.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the suMN of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb417['A'].sum(),
        'C': peaksdb417['C'].sum(),
        'D': peaksdb417['D'].sum(),
        'E': peaksdb417['E'].sum(),
        'F': peaksdb417['F'].sum(),
        'G': peaksdb417['G'].sum(),
        'H': peaksdb417['H'].sum(),
        'I': peaksdb417['I'].sum(),
        'K': peaksdb417['K'].sum(),
        'L': peaksdb417['L'].sum(),
        'M': peaksdb417['M'].sum(),
        'N': peaksdb417['N'].sum(),
        'P': peaksdb417['P'].sum(),
        'Q': peaksdb417['Q'].sum(),
        'R': peaksdb417['R'].sum(),
        'S': peaksdb417['S'].sum(),
        'T': peaksdb417['T'].sum(),
        'V': peaksdb417['V'].sum(),
        'W': peaksdb417['W'].sum(),
        'Y': peaksdb417['Y'].sum(),
        'c-carb': peaksdb417['c-carb'].sum(),
        'm-oxid': peaksdb417['m-oxid'].sum(),
        'n-deam': peaksdb417['n-deam'].sum(),
        'q-deam': peaksdb417['q-deam'].sum(),
        'Total area': peaksdb417['Area'].sum(),
        'Total length': peaksdb417['stripped_length'].sum()
       }

totalpeaksdb417 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 
                                              'n-deam', 'q-deam', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb417['% C w/ carb'] = totalpeaksdb417['c-carb'] / totalpeaksdb417['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb417['% M w/ oxid'] = totalpeaksdb417['m-oxid'] / totalpeaksdb417['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb417['% N w/ deam'] = totalpeaksdb417['n-deam'] / totalpeaksdb417['N'] 

# calculate percentage of N's that are deamidated
totalpeaksdb417['% Q w/ deam'] = totalpeaksdb417['q-deam'] / totalpeaksdb417['Q'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb417['NAAF denom.'] = totalpeaksdb417['Total area'] / totalpeaksdb417['Total length']

# write modified dataframe to new txt file
totalpeaksdb417.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/PeaksDB/417_MN_T24_GD_PDB/417B_MN_T24_GD_PDB_totals.csv")

totalpeaksdb417.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb417 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 511882.681055

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb417['NAAF factor'] = (peaksdb417['NAAF_num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMN
peaksdb417_NAAF = peaksdb417[['stripped_peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'n-deam', 'q-deam']].copy()

# multiply the NAAF20 factor by the AAs to normalize its abundance by peak area and peptide length

peaksdb417_NAAF['A-NAAF20'] = peaksdb417_NAAF['A'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['C-NAAF20'] = peaksdb417_NAAF['C'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['D-NAAF20'] = peaksdb417_NAAF['D'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['E-NAAF20'] = peaksdb417_NAAF['E'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['F-NAAF20'] = peaksdb417_NAAF['F'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['G-NAAF20'] = peaksdb417_NAAF['G'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['H-NAAF20'] = peaksdb417_NAAF['H'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['I-NAAF20'] = peaksdb417_NAAF['I'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['K-NAAF20'] = peaksdb417_NAAF['K'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['L-NAAF20'] = peaksdb417_NAAF['L'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['M-NAAF20'] = peaksdb417_NAAF['M'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['N-NAAF20'] = peaksdb417_NAAF['N'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['P-NAAF20'] = peaksdb417_NAAF['P'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['Q-NAAF20'] = peaksdb417_NAAF['Q'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['R-NAAF20'] = peaksdb417_NAAF['R'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['S-NAAF20'] = peaksdb417_NAAF['S'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['T-NAAF20'] = peaksdb417_NAAF['T'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['V-NAAF20'] = peaksdb417_NAAF['V'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['W-NAAF20'] = peaksdb417_NAAF['W'] * peaksdb417['NAAF factor']
peaksdb417_NAAF['Y-NAAF20'] = peaksdb417_NAAF['Y'] * peaksdb417['NAAF factor']

# multiply the NAAF20 factor by the PTMN normalize its abundance by peak area and peptide length

peaksdb417_NAAF['ccarb-NAAF20'] = peaksdb417_NAAF['c-carb'] * peaksdb417_NAAF['NAAF factor']
peaksdb417_NAAF['moxid-NAAF20'] = peaksdb417_NAAF['m-oxid'] * peaksdb417_NAAF['NAAF factor']
peaksdb417_NAAF['ndeam-NAAF20'] = peaksdb417_NAAF['n-deam'] * peaksdb417_NAAF['NAAF factor']
peaksdb417_NAAF['qdeam-NAAF20'] = peaksdb417_NAAF['q-deam'] * peaksdb417_NAAF['NAAF factor']


# write the dataframe to a new csv
peaksdb417_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/PeaksDB/417_MN_T24_GD_PDB/417B_MN_T24_GD_PDB_naaf.csv")

peaksdb417_NAAF.head()
```

```python
# made a new dataframe that contains the suMN of NAAF normalized AAs for peaksdb417 peaksdb results
# also contains the suMN of the NAAF-corrected PTMN occurances for each affected residue

index = ['sample total']

data = {'NAAF': peaksdb417_NAAF['NAAF factor'].sum(),
        'A': peaksdb417_NAAF['A-NAAF20'].sum(),
        'C': peaksdb417_NAAF['C-NAAF20'].sum(),
        'D': peaksdb417_NAAF['D-NAAF20'].sum(),
        'E': peaksdb417_NAAF['E-NAAF20'].sum(),
        'F': peaksdb417_NAAF['F-NAAF20'].sum(),
        'G': peaksdb417_NAAF['G-NAAF20'].sum(),
        'H': peaksdb417_NAAF['H-NAAF20'].sum(),
        'I': peaksdb417_NAAF['I-NAAF20'].sum(),
        'K': peaksdb417_NAAF['K-NAAF20'].sum(),
        'L': peaksdb417_NAAF['L-NAAF20'].sum(),
        'M': peaksdb417_NAAF['M-NAAF20'].sum(),
        'N': peaksdb417_NAAF['N-NAAF20'].sum(),
        'P': peaksdb417_NAAF['P-NAAF20'].sum(),
        'Q': peaksdb417_NAAF['Q-NAAF20'].sum(),
        'R': peaksdb417_NAAF['R-NAAF20'].sum(),
        'S': peaksdb417_NAAF['S-NAAF20'].sum(),
        'T': peaksdb417_NAAF['T-NAAF20'].sum(),
        'V': peaksdb417_NAAF['V-NAAF20'].sum(),
        'W': peaksdb417_NAAF['W-NAAF20'].sum(),
        'Y': peaksdb417_NAAF['Y-NAAF20'].sum(),
        'c-carb': peaksdb417_NAAF['ccarb-NAAF20'].sum(),
        'm-oxid': peaksdb417_NAAF['moxid-NAAF20'].sum(),
        'n-deam': peaksdb417_NAAF['ndeam-NAAF20'].sum(),
        'q-deam': peaksdb417_NAAF['qdeam-NAAF20'].sum()
       }

totalpeaksdb417_NAAF = pd.DataFrame(data, columns=['NAAF', 'A', 'C', 'D', 'E', 'F', \
                                                   'G', 'H', 'I','K', 'L', 'M', \
                                                   'N', 'P', 'Q', 'R', 'S', \
                                                   'T', 'V', 'W', 'Y', 'c-carb', \
                                                   'm-oxid', 'n-deam', 'q-deam'], index=index)

# calculate the NAAF-corrected % modified C, M, N, Q, K, and Rs


# calculate NAAF-corrected percentage of C's with carb (should be 1.0)
totalpeaksdb417_NAAF['% C w/ carb.'] = totalpeaksdb417_NAAF['c-carb'] / totalpeaksdb417_NAAF['C'] 

# calculate NAAF-corrected percentage of M's that are oxidized
totalpeaksdb417_NAAF['% M w/ oxid'] = totalpeaksdb417_NAAF['m-oxid'] / totalpeaksdb417_NAAF['M'] 

# calculate NAAF-corrected percentage of N's that are deamidated
totalpeaksdb417_NAAF['% N w/ deam'] = totalpeaksdb417_NAAF['n-deam'] / totalpeaksdb417_NAAF['N'] 

# calculate NAAF-corrected percentage of N's that are deamidated
totalpeaksdb417_NAAF['% Q w/ deam'] = totalpeaksdb417_NAAF['q-deam'] / totalpeaksdb417_NAAF['Q'] 

# calculate NAAF summed numerator over denominator (in above cell) for all peptides in dataset i: a check
totalpeaksdb417_NAAF['NAAF check'] = totalpeaksdb417_NAAF['NAAF'] / 511882.681055

# write modified dataframe to new txt file, same name + totals
totalpeaksdb417_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas7-incubations/processed/PeaksDB/417_MN_T24_GD_PDB/417B_MN_T24_GD_PDB_naaf_totals.csv")

totalpeaksdb417_NAAF.head()
```
