### Manipulation of PeaskDB de novo-assisted database search results of Trocas 8 (low water, November 2019) non incubation samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the TROCAS # (e.g., 675) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup689 = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/T8-June2021-UWPR-DN50_10lgP15/689_Trocas8_June2021_PEAKS_149/689_Trocas8_June2021_PEAKS_149_peptide.csv")

# remove redundant rows
peaksdb689 = pd.DataFrame.drop_duplicates(peaksdbdup689)

print(peaksdb689.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source_File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb689.columns = columns

#remmove # spec and accession columns because they mess parsing up

del peaksdb689['#Spec']
del peaksdb689['Accession']
del peaksdb689['PTM']
del peaksdb689['AScore']

mean_length = peaksdb689['Length'].mean()
print(mean_length)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup689))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb689))

#look at the dataframe
peaksdb689.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb689['A'] = peaksdb689['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb689['C'] = peaksdb689['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb689['D'] = peaksdb689['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb689['E'] = peaksdb689['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb689['F'] = peaksdb689['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb689['G'] = peaksdb689['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb689['H'] = peaksdb689['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb689 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb689['I'] = peaksdb689['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb689['K'] = peaksdb689['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb689['L'] = peaksdb689['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb689['M'] = peaksdb689['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb689['N'] = peaksdb689['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb689['P'] = peaksdb689['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb689['Q'] = peaksdb689['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb689['R'] = peaksdb689['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb689['S'] = peaksdb689['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb689['T'] = peaksdb689['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb689['V'] = peaksdb689['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb689['W'] = peaksdb689['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb689['Y'] = peaksdb689['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb689['c-carb'] = peaksdb689['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb689['m-oxid'] = peaksdb689['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb689['n-deam'] = peaksdb689['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a lamba function to enumerate the # of deamidated Q's in each peptide
peaksdb689['q-deam'] = peaksdb689['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb689['stripped_peptide'] = peaksdb689['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb689['stripped_length'] = peaksdb689['stripped_peptide'].apply(len)

peaksdb689['NAAF_num.'] = peaksdb689['Area'] / peaksdb689['stripped_length']

# total the number of modifications in sequence
peaksdb689['ptm-total'] = peaksdb689['c-carb'] + peaksdb689['m-oxid'] + peaksdb689['n-deam'] + peaksdb689['q-deam']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb689['stripped_IL']= peaksdb689['stripped_peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb689.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/681_NMCP_50_GF75_PDB/689_NMCP_50_GF75_PDB.csv")

# check out the results
peaksdb689.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb689['A'].sum(),
        'C': peaksdb689['C'].sum(),
        'D': peaksdb689['D'].sum(),
        'E': peaksdb689['E'].sum(),
        'F': peaksdb689['F'].sum(),
        'G': peaksdb689['G'].sum(),
        'H': peaksdb689['H'].sum(),
        'I': peaksdb689['I'].sum(),
        'K': peaksdb689['K'].sum(),
        'L': peaksdb689['L'].sum(),
        'M': peaksdb689['M'].sum(),
        'N': peaksdb689['N'].sum(),
        'P': peaksdb689['P'].sum(),
        'Q': peaksdb689['Q'].sum(),
        'R': peaksdb689['R'].sum(),
        'S': peaksdb689['S'].sum(),
        'T': peaksdb689['T'].sum(),
        'V': peaksdb689['V'].sum(),
        'W': peaksdb689['W'].sum(),
        'Y': peaksdb689['Y'].sum(),
        'c-carb': peaksdb689['c-carb'].sum(),
        'm-oxid': peaksdb689['m-oxid'].sum(),
        'n-deam': peaksdb689['n-deam'].sum(),
        'q-deam': peaksdb689['q-deam'].sum(),
        'Total area': peaksdb689['Area'].sum(),
        'Total length': peaksdb689['stripped_length'].sum()
       }

totalpeaksdb689 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 
                                              'n-deam', 'q-deam', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb689['% C w/ carb'] = totalpeaksdb689['c-carb'] / totalpeaksdb689['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb689['% M w/ oxid'] = totalpeaksdb689['m-oxid'] / totalpeaksdb689['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb689['% N w/ deam'] = totalpeaksdb689['n-deam'] / totalpeaksdb689['N'] 

# calculate percentage of N's that are deamidated
totalpeaksdb689['% Q w/ deam'] = totalpeaksdb689['q-deam'] / totalpeaksdb689['Q'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb689['NAAF denom.'] = totalpeaksdb689['Total area'] / totalpeaksdb689['Total length']

# write modified dataframe to new txt file
totalpeaksdb689.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/681_NMCP_50_GF75_PDB/689_NMCP_50_GF75_PDB_totals.csv")

totalpeaksdb689.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb689 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 266103.038278

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb689['NAAF factor'] = (peaksdb689['NAAF_num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb689_NAAF = peaksdb689[['stripped_peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'n-deam', 'q-deam']].copy()

# multiply the NAAF20 factor NMCP the AAs to normalize its abundance NMCP peak area and peptide length

peaksdb689_NAAF['A-NAAF20'] = peaksdb689_NAAF['A'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['C-NAAF20'] = peaksdb689_NAAF['C'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['D-NAAF20'] = peaksdb689_NAAF['D'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['E-NAAF20'] = peaksdb689_NAAF['E'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['F-NAAF20'] = peaksdb689_NAAF['F'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['G-NAAF20'] = peaksdb689_NAAF['G'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['H-NAAF20'] = peaksdb689_NAAF['H'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['I-NAAF20'] = peaksdb689_NAAF['I'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['K-NAAF20'] = peaksdb689_NAAF['K'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['L-NAAF20'] = peaksdb689_NAAF['L'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['M-NAAF20'] = peaksdb689_NAAF['M'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['N-NAAF20'] = peaksdb689_NAAF['N'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['P-NAAF20'] = peaksdb689_NAAF['P'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['Q-NAAF20'] = peaksdb689_NAAF['Q'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['R-NAAF20'] = peaksdb689_NAAF['R'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['S-NAAF20'] = peaksdb689_NAAF['S'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['T-NAAF20'] = peaksdb689_NAAF['T'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['V-NAAF20'] = peaksdb689_NAAF['V'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['W-NAAF20'] = peaksdb689_NAAF['W'] * peaksdb689['NAAF factor']
peaksdb689_NAAF['Y-NAAF20'] = peaksdb689_NAAF['Y'] * peaksdb689['NAAF factor']

# multiply the NAAF20 factor NMCP the PTMs normalize its abundance NMCP peak area and peptide length

peaksdb689_NAAF['ccarb-NAAF20'] = peaksdb689_NAAF['c-carb'] * peaksdb689_NAAF['NAAF factor']
peaksdb689_NAAF['moxid-NAAF20'] = peaksdb689_NAAF['m-oxid'] * peaksdb689_NAAF['NAAF factor']
peaksdb689_NAAF['ndeam-NAAF20'] = peaksdb689_NAAF['n-deam'] * peaksdb689_NAAF['NAAF factor']
peaksdb689_NAAF['qdeam-NAAF20'] = peaksdb689_NAAF['q-deam'] * peaksdb689_NAAF['NAAF factor']


# write the dataframe to a new csv
peaksdb689_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/681_NMCP_50_GF75_PDB/689_NMCP_50_GF75_PDB_naaf.csv")

peaksdb689_NAAF.head()
```
