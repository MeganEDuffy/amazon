### Manipulation of PeaskDB de novo-assisted database search results of Trocas 8 (low water, November 2019) non incubation samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the TROCAS # (e.g., 675) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup677 = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/T8-June2021-UWPR-DN50_10lgP15/677_Trocas8_June2021_PEAKS_135/677_Trocas8_June2021_PEAKS_135_peptide.csv")

# remove redundant rows
peaksdb677 = pd.DataFrame.drop_duplicates(peaksdbdup677)

print(peaksdb677.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source_File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb677.columns = columns

#remmove # spec and accession columns because they mess parsing up

del peaksdb677['#Spec']
del peaksdb677['Accession']
del peaksdb677['PTM']
del peaksdb677['AScore']

mean_length = peaksdb677['Length'].mean()
print(mean_length)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup677))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb677))

#look at the dataframe
peaksdb677.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb677['A'] = peaksdb677['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb677['C'] = peaksdb677['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb677['D'] = peaksdb677['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb677['E'] = peaksdb677['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb677['F'] = peaksdb677['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb677['G'] = peaksdb677['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb677['H'] = peaksdb677['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb677 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb677['I'] = peaksdb677['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb677['K'] = peaksdb677['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb677['L'] = peaksdb677['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb677['M'] = peaksdb677['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb677['N'] = peaksdb677['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb677['P'] = peaksdb677['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb677['Q'] = peaksdb677['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb677['R'] = peaksdb677['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb677['S'] = peaksdb677['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb677['T'] = peaksdb677['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb677['V'] = peaksdb677['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb677['W'] = peaksdb677['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb677['Y'] = peaksdb677['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb677['c-carb'] = peaksdb677['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb677['m-oxid'] = peaksdb677['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb677['n-deam'] = peaksdb677['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a lamba function to enumerate the # of deamidated Q's in each peptide
peaksdb677['q-deam'] = peaksdb677['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb677['stripped_peptide'] = peaksdb677['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb677['stripped_length'] = peaksdb677['stripped_peptide'].apply(len)

peaksdb677['NAAF_num.'] = peaksdb677['Area'] / peaksdb677['stripped_length']

# total the number of modifications in sequence
peaksdb677['ptm-total'] = peaksdb677['c-carb'] + peaksdb677['m-oxid'] + peaksdb677['n-deam'] + peaksdb677['q-deam']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb677['stripped_IL']= peaksdb677['stripped_peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb677.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/677_CV_50_GF75_PDB/677_CV_50_GF75_PDB.csv")

# check out the results
peaksdb677.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb677['A'].sum(),
        'C': peaksdb677['C'].sum(),
        'D': peaksdb677['D'].sum(),
        'E': peaksdb677['E'].sum(),
        'F': peaksdb677['F'].sum(),
        'G': peaksdb677['G'].sum(),
        'H': peaksdb677['H'].sum(),
        'I': peaksdb677['I'].sum(),
        'K': peaksdb677['K'].sum(),
        'L': peaksdb677['L'].sum(),
        'M': peaksdb677['M'].sum(),
        'N': peaksdb677['N'].sum(),
        'P': peaksdb677['P'].sum(),
        'Q': peaksdb677['Q'].sum(),
        'R': peaksdb677['R'].sum(),
        'S': peaksdb677['S'].sum(),
        'T': peaksdb677['T'].sum(),
        'V': peaksdb677['V'].sum(),
        'W': peaksdb677['W'].sum(),
        'Y': peaksdb677['Y'].sum(),
        'c-carb': peaksdb677['c-carb'].sum(),
        'm-oxid': peaksdb677['m-oxid'].sum(),
        'n-deam': peaksdb677['n-deam'].sum(),
        'q-deam': peaksdb677['q-deam'].sum(),
        'Total area': peaksdb677['Area'].sum(),
        'Total length': peaksdb677['stripped_length'].sum()
       }

totalpeaksdb677 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 
                                              'n-deam', 'q-deam', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb677['% C w/ carb'] = totalpeaksdb677['c-carb'] / totalpeaksdb677['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb677['% M w/ oxid'] = totalpeaksdb677['m-oxid'] / totalpeaksdb677['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb677['% N w/ deam'] = totalpeaksdb677['n-deam'] / totalpeaksdb677['N'] 

# calculate percentage of N's that are deamidated
totalpeaksdb677['% Q w/ deam'] = totalpeaksdb677['q-deam'] / totalpeaksdb677['Q'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb677['NAAF denom.'] = totalpeaksdb677['Total area'] / totalpeaksdb677['Total length']

# write modified dataframe to new txt file
totalpeaksdb677.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/677_CV_50_GF75_PDB/677_CV_50_GF75_PDB_totals.csv")

totalpeaksdb677.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb677 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 123361.316901

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb677['NAAF factor'] = (peaksdb677['NAAF_num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb677_NAAF = peaksdb677[['stripped_peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'n-deam', 'q-deam']].copy()

# multiply the NAAF20 factor CV the AAs to normalize its abundance CV peak area and peptide length

peaksdb677_NAAF['A-NAAF20'] = peaksdb677_NAAF['A'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['C-NAAF20'] = peaksdb677_NAAF['C'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['D-NAAF20'] = peaksdb677_NAAF['D'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['E-NAAF20'] = peaksdb677_NAAF['E'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['F-NAAF20'] = peaksdb677_NAAF['F'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['G-NAAF20'] = peaksdb677_NAAF['G'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['H-NAAF20'] = peaksdb677_NAAF['H'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['I-NAAF20'] = peaksdb677_NAAF['I'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['K-NAAF20'] = peaksdb677_NAAF['K'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['L-NAAF20'] = peaksdb677_NAAF['L'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['M-NAAF20'] = peaksdb677_NAAF['M'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['N-NAAF20'] = peaksdb677_NAAF['N'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['P-NAAF20'] = peaksdb677_NAAF['P'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['Q-NAAF20'] = peaksdb677_NAAF['Q'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['R-NAAF20'] = peaksdb677_NAAF['R'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['S-NAAF20'] = peaksdb677_NAAF['S'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['T-NAAF20'] = peaksdb677_NAAF['T'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['V-NAAF20'] = peaksdb677_NAAF['V'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['W-NAAF20'] = peaksdb677_NAAF['W'] * peaksdb677['NAAF factor']
peaksdb677_NAAF['Y-NAAF20'] = peaksdb677_NAAF['Y'] * peaksdb677['NAAF factor']

# multiply the NAAF20 factor CV the PTMs normalize its abundance CV peak area and peptide length

peaksdb677_NAAF['ccarb-NAAF20'] = peaksdb677_NAAF['c-carb'] * peaksdb677_NAAF['NAAF factor']
peaksdb677_NAAF['moxid-NAAF20'] = peaksdb677_NAAF['m-oxid'] * peaksdb677_NAAF['NAAF factor']
peaksdb677_NAAF['ndeam-NAAF20'] = peaksdb677_NAAF['n-deam'] * peaksdb677_NAAF['NAAF factor']
peaksdb677_NAAF['qdeam-NAAF20'] = peaksdb677_NAAF['q-deam'] * peaksdb677_NAAF['NAAF factor']


# write the dataframe to a new csv
peaksdb677_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/677_CV_50_GF75_PDB/677_CV_50_GF75_PDB_naaf.csv")

peaksdb677_NAAF.head()
```
