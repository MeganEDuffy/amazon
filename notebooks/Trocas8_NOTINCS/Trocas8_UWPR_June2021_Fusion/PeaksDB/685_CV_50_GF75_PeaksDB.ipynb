### Manipulation of PeaskDB de novo-assisted database search results of Trocas 8 (low water, November 2019) non incubation samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the TROCAS # (e.g., 675) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup685 = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/T8-June2021-UWPR-DN50_10lgP15/685_Trocas8_June2021_PEAKS_144/685_Trocas8_June2021_PEAKS_144_peptide.csv")

# remove redundant rows
peaksdb685 = pd.DataFrame.drop_duplicates(peaksdbdup685)

print(peaksdb685.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source_File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb685.columns = columns

#remmove # spec and accession columns because they mess parsing up

del peaksdb685['#Spec']
del peaksdb685['Accession']
del peaksdb685['PTM']
del peaksdb685['AScore']

mean_length = peaksdb685['Length'].mean()
print(mean_length)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup685))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb685))

#look at the dataframe
peaksdb685.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb685['A'] = peaksdb685['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb685['C'] = peaksdb685['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb685['D'] = peaksdb685['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb685['E'] = peaksdb685['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb685['F'] = peaksdb685['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb685['G'] = peaksdb685['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb685['H'] = peaksdb685['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb685 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb685['I'] = peaksdb685['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb685['K'] = peaksdb685['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb685['L'] = peaksdb685['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb685['M'] = peaksdb685['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb685['N'] = peaksdb685['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb685['P'] = peaksdb685['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb685['Q'] = peaksdb685['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb685['R'] = peaksdb685['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb685['S'] = peaksdb685['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb685['T'] = peaksdb685['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb685['V'] = peaksdb685['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb685['W'] = peaksdb685['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb685['Y'] = peaksdb685['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb685['c-carb'] = peaksdb685['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb685['m-oxid'] = peaksdb685['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb685['n-deam'] = peaksdb685['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a lamba function to enumerate the # of deamidated Q's in each peptide
peaksdb685['q-deam'] = peaksdb685['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb685['stripped_peptide'] = peaksdb685['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb685['stripped_length'] = peaksdb685['stripped_peptide'].apply(len)

peaksdb685['NAAF_num.'] = peaksdb685['Area'] / peaksdb685['stripped_length']

# total the number of modifications in sequence
peaksdb685['ptm-total'] = peaksdb685['c-carb'] + peaksdb685['m-oxid'] + peaksdb685['n-deam'] + peaksdb685['q-deam']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb685['stripped_IL']= peaksdb685['stripped_peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb685.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/677_CV_50_GF75_PDB/685_CV_50_GF75_PDB.csv")

# check out the results
peaksdb685.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb685['A'].sum(),
        'C': peaksdb685['C'].sum(),
        'D': peaksdb685['D'].sum(),
        'E': peaksdb685['E'].sum(),
        'F': peaksdb685['F'].sum(),
        'G': peaksdb685['G'].sum(),
        'H': peaksdb685['H'].sum(),
        'I': peaksdb685['I'].sum(),
        'K': peaksdb685['K'].sum(),
        'L': peaksdb685['L'].sum(),
        'M': peaksdb685['M'].sum(),
        'N': peaksdb685['N'].sum(),
        'P': peaksdb685['P'].sum(),
        'Q': peaksdb685['Q'].sum(),
        'R': peaksdb685['R'].sum(),
        'S': peaksdb685['S'].sum(),
        'T': peaksdb685['T'].sum(),
        'V': peaksdb685['V'].sum(),
        'W': peaksdb685['W'].sum(),
        'Y': peaksdb685['Y'].sum(),
        'c-carb': peaksdb685['c-carb'].sum(),
        'm-oxid': peaksdb685['m-oxid'].sum(),
        'n-deam': peaksdb685['n-deam'].sum(),
        'q-deam': peaksdb685['q-deam'].sum(),
        'Total area': peaksdb685['Area'].sum(),
        'Total length': peaksdb685['stripped_length'].sum()
       }

totalpeaksdb685 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 
                                              'n-deam', 'q-deam', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb685['% C w/ carb'] = totalpeaksdb685['c-carb'] / totalpeaksdb685['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb685['% M w/ oxid'] = totalpeaksdb685['m-oxid'] / totalpeaksdb685['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb685['% N w/ deam'] = totalpeaksdb685['n-deam'] / totalpeaksdb685['N'] 

# calculate percentage of N's that are deamidated
totalpeaksdb685['% Q w/ deam'] = totalpeaksdb685['q-deam'] / totalpeaksdb685['Q'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb685['NAAF denom.'] = totalpeaksdb685['Total area'] / totalpeaksdb685['Total length']

# write modified dataframe to new txt file
totalpeaksdb685.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/677_CV_50_GF75_PDB/685_CV_50_GF75_PDB_totals.csv")

totalpeaksdb685.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb685 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 167928.354978

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb685['NAAF factor'] = (peaksdb685['NAAF_num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb685_NAAF = peaksdb685[['stripped_peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'n-deam', 'q-deam']].copy()

# multiply the NAAF20 factor CV the AAs to normalize its abundance CV peak area and peptide length

peaksdb685_NAAF['A-NAAF20'] = peaksdb685_NAAF['A'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['C-NAAF20'] = peaksdb685_NAAF['C'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['D-NAAF20'] = peaksdb685_NAAF['D'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['E-NAAF20'] = peaksdb685_NAAF['E'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['F-NAAF20'] = peaksdb685_NAAF['F'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['G-NAAF20'] = peaksdb685_NAAF['G'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['H-NAAF20'] = peaksdb685_NAAF['H'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['I-NAAF20'] = peaksdb685_NAAF['I'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['K-NAAF20'] = peaksdb685_NAAF['K'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['L-NAAF20'] = peaksdb685_NAAF['L'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['M-NAAF20'] = peaksdb685_NAAF['M'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['N-NAAF20'] = peaksdb685_NAAF['N'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['P-NAAF20'] = peaksdb685_NAAF['P'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['Q-NAAF20'] = peaksdb685_NAAF['Q'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['R-NAAF20'] = peaksdb685_NAAF['R'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['S-NAAF20'] = peaksdb685_NAAF['S'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['T-NAAF20'] = peaksdb685_NAAF['T'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['V-NAAF20'] = peaksdb685_NAAF['V'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['W-NAAF20'] = peaksdb685_NAAF['W'] * peaksdb685['NAAF factor']
peaksdb685_NAAF['Y-NAAF20'] = peaksdb685_NAAF['Y'] * peaksdb685['NAAF factor']

# multiply the NAAF20 factor CV the PTMs normalize its abundance CV peak area and peptide length

peaksdb685_NAAF['ccarb-NAAF20'] = peaksdb685_NAAF['c-carb'] * peaksdb685_NAAF['NAAF factor']
peaksdb685_NAAF['moxid-NAAF20'] = peaksdb685_NAAF['m-oxid'] * peaksdb685_NAAF['NAAF factor']
peaksdb685_NAAF['ndeam-NAAF20'] = peaksdb685_NAAF['n-deam'] * peaksdb685_NAAF['NAAF factor']
peaksdb685_NAAF['qdeam-NAAF20'] = peaksdb685_NAAF['q-deam'] * peaksdb685_NAAF['NAAF factor']


# write the dataframe to a new csv
peaksdb685_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/677_CV_50_GF75_PDB/685_CV_50_GF75_PDB_naaf.csv")

peaksdb685_NAAF.head()
```
