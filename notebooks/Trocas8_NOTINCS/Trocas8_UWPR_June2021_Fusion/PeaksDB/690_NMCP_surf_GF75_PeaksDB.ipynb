### Manipulation of PeaskDB de novo-assisted database search results of Trocas 8 (low water, November 2019) non incubation samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the TROCAS # (e.g., 675) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup690 = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/T8-June2021-UWPR-DN50_10lgP15/690_Trocas8_June2021_PEAKS_151/690_Trocas8_June2021_PEAKS_151_peptide.csv")

# remove redundant rows
peaksdb690 = pd.DataFrame.drop_duplicates(peaksdbdup690)

print(peaksdb690.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source_File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb690.columns = columns

#remmove # spec and accession columns because they mess parsing up

del peaksdb690['#Spec']
del peaksdb690['Accession']
del peaksdb690['PTM']
del peaksdb690['AScore']

mean_length = peaksdb690['Length'].mean()
print(mean_length)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup690))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb690))

#look at the dataframe
peaksdb690.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb690['A'] = peaksdb690['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb690['C'] = peaksdb690['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb690['D'] = peaksdb690['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb690['E'] = peaksdb690['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb690['F'] = peaksdb690['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb690['G'] = peaksdb690['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb690['H'] = peaksdb690['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb690 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb690['I'] = peaksdb690['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb690['K'] = peaksdb690['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb690['L'] = peaksdb690['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb690['M'] = peaksdb690['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb690['N'] = peaksdb690['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb690['P'] = peaksdb690['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb690['Q'] = peaksdb690['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb690['R'] = peaksdb690['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb690['S'] = peaksdb690['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb690['T'] = peaksdb690['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb690['V'] = peaksdb690['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb690['W'] = peaksdb690['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb690['Y'] = peaksdb690['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb690['c-carb'] = peaksdb690['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb690['m-oxid'] = peaksdb690['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb690['n-deam'] = peaksdb690['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a lamba function to enumerate the # of deamidated Q's in each peptide
peaksdb690['q-deam'] = peaksdb690['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb690['stripped_peptide'] = peaksdb690['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb690['stripped_length'] = peaksdb690['stripped_peptide'].apply(len)

peaksdb690['NAAF_num.'] = peaksdb690['Area'] / peaksdb690['stripped_length']

# total the number of modifications in sequence
peaksdb690['ptm-total'] = peaksdb690['c-carb'] + peaksdb690['m-oxid'] + peaksdb690['n-deam'] + peaksdb690['q-deam']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb690['stripped_IL']= peaksdb690['stripped_peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb690.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/682_NMCP_surf_GF75_PDB/690_NMCP_surf_GF75_PDB.csv")

# check out the results
peaksdb690.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb690['A'].sum(),
        'C': peaksdb690['C'].sum(),
        'D': peaksdb690['D'].sum(),
        'E': peaksdb690['E'].sum(),
        'F': peaksdb690['F'].sum(),
        'G': peaksdb690['G'].sum(),
        'H': peaksdb690['H'].sum(),
        'I': peaksdb690['I'].sum(),
        'K': peaksdb690['K'].sum(),
        'L': peaksdb690['L'].sum(),
        'M': peaksdb690['M'].sum(),
        'N': peaksdb690['N'].sum(),
        'P': peaksdb690['P'].sum(),
        'Q': peaksdb690['Q'].sum(),
        'R': peaksdb690['R'].sum(),
        'S': peaksdb690['S'].sum(),
        'T': peaksdb690['T'].sum(),
        'V': peaksdb690['V'].sum(),
        'W': peaksdb690['W'].sum(),
        'Y': peaksdb690['Y'].sum(),
        'c-carb': peaksdb690['c-carb'].sum(),
        'm-oxid': peaksdb690['m-oxid'].sum(),
        'n-deam': peaksdb690['n-deam'].sum(),
        'q-deam': peaksdb690['q-deam'].sum(),
        'Total area': peaksdb690['Area'].sum(),
        'Total length': peaksdb690['stripped_length'].sum()
       }

totalpeaksdb690 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 
                                              'n-deam', 'q-deam', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb690['% C w/ carb'] = totalpeaksdb690['c-carb'] / totalpeaksdb690['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb690['% M w/ oxid'] = totalpeaksdb690['m-oxid'] / totalpeaksdb690['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb690['% N w/ deam'] = totalpeaksdb690['n-deam'] / totalpeaksdb690['N'] 

# calculate percentage of N's that are deamidated
totalpeaksdb690['% Q w/ deam'] = totalpeaksdb690['q-deam'] / totalpeaksdb690['Q'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb690['NAAF denom.'] = totalpeaksdb690['Total area'] / totalpeaksdb690['Total length']

# write modified dataframe to new txt file
totalpeaksdb690.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/682_NMCP_surf_GF75_PDB/690_NMCP_surf_GF75_PDB_totals.csv")

totalpeaksdb690.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb690 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 208107.60396

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb690['NAAF factor'] = (peaksdb690['NAAF_num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb690_NAAF = peaksdb690[['stripped_peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'n-deam', 'q-deam']].copy()

# multiply the NAAF20 factor NMCP the AAs to normalize its abundance NMCP peak area and peptide length

peaksdb690_NAAF['A-NAAF20'] = peaksdb690_NAAF['A'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['C-NAAF20'] = peaksdb690_NAAF['C'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['D-NAAF20'] = peaksdb690_NAAF['D'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['E-NAAF20'] = peaksdb690_NAAF['E'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['F-NAAF20'] = peaksdb690_NAAF['F'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['G-NAAF20'] = peaksdb690_NAAF['G'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['H-NAAF20'] = peaksdb690_NAAF['H'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['I-NAAF20'] = peaksdb690_NAAF['I'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['K-NAAF20'] = peaksdb690_NAAF['K'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['L-NAAF20'] = peaksdb690_NAAF['L'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['M-NAAF20'] = peaksdb690_NAAF['M'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['N-NAAF20'] = peaksdb690_NAAF['N'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['P-NAAF20'] = peaksdb690_NAAF['P'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['Q-NAAF20'] = peaksdb690_NAAF['Q'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['R-NAAF20'] = peaksdb690_NAAF['R'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['S-NAAF20'] = peaksdb690_NAAF['S'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['T-NAAF20'] = peaksdb690_NAAF['T'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['V-NAAF20'] = peaksdb690_NAAF['V'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['W-NAAF20'] = peaksdb690_NAAF['W'] * peaksdb690['NAAF factor']
peaksdb690_NAAF['Y-NAAF20'] = peaksdb690_NAAF['Y'] * peaksdb690['NAAF factor']

# multiply the NAAF20 factor NMCP the PTMs normalize its abundance NMCP peak area and peptide length

peaksdb690_NAAF['ccarb-NAAF20'] = peaksdb690_NAAF['c-carb'] * peaksdb690_NAAF['NAAF factor']
peaksdb690_NAAF['moxid-NAAF20'] = peaksdb690_NAAF['m-oxid'] * peaksdb690_NAAF['NAAF factor']
peaksdb690_NAAF['ndeam-NAAF20'] = peaksdb690_NAAF['n-deam'] * peaksdb690_NAAF['NAAF factor']
peaksdb690_NAAF['qdeam-NAAF20'] = peaksdb690_NAAF['q-deam'] * peaksdb690_NAAF['NAAF factor']


# write the dataframe to a new csv
peaksdb690_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/682_NMCP_surf_GF75_PDB/690_NMCP_surf_GF75_PDB_naaf.csv")

peaksdb690_NAAF.head()
```
