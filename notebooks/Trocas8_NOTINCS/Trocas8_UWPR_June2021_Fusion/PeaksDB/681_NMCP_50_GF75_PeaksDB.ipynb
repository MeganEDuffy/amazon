### Manipulation of PeaskDB de novo-assisted database search results of Trocas 8 (low water, November 2019) non incubation samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the TROCAS # (e.g., 675) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup681 = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/T8-June2021-UWPR-DN50_10lgP15/681_Trocas8_June2021_PEAKS_140/681_Trocas8_June2021_PEAKS_140_peptide.csv")

# remove redundant rows
peaksdb681 = pd.DataFrame.drop_duplicates(peaksdbdup681)

print(peaksdb681.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source_File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb681.columns = columns

#remmove # spec and accession columns because they mess parsing up

del peaksdb681['#Spec']
del peaksdb681['Accession']
del peaksdb681['PTM']
del peaksdb681['AScore']

mean_length = peaksdb681['Length'].mean()
print(mean_length)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup681))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb681))

#look at the dataframe
peaksdb681.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb681['A'] = peaksdb681['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb681['C'] = peaksdb681['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb681['D'] = peaksdb681['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb681['E'] = peaksdb681['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb681['F'] = peaksdb681['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb681['G'] = peaksdb681['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb681['H'] = peaksdb681['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb681 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb681['I'] = peaksdb681['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb681['K'] = peaksdb681['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb681['L'] = peaksdb681['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb681['M'] = peaksdb681['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb681['N'] = peaksdb681['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb681['P'] = peaksdb681['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb681['Q'] = peaksdb681['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb681['R'] = peaksdb681['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb681['S'] = peaksdb681['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb681['T'] = peaksdb681['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb681['V'] = peaksdb681['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb681['W'] = peaksdb681['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb681['Y'] = peaksdb681['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb681['c-carb'] = peaksdb681['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb681['m-oxid'] = peaksdb681['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb681['n-deam'] = peaksdb681['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a lamba function to enumerate the # of deamidated Q's in each peptide
peaksdb681['q-deam'] = peaksdb681['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb681['stripped_peptide'] = peaksdb681['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb681['stripped_length'] = peaksdb681['stripped_peptide'].apply(len)

peaksdb681['NAAF_num.'] = peaksdb681['Area'] / peaksdb681['stripped_length']

# total the number of modifications in sequence
peaksdb681['ptm-total'] = peaksdb681['c-carb'] + peaksdb681['m-oxid'] + peaksdb681['n-deam'] + peaksdb681['q-deam']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb681['stripped_IL']= peaksdb681['stripped_peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb681.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/681_NMCP_50_GF75_PDB/681_NMCP_50_GF75_PDB.csv")

# check out the results
peaksdb681.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb681['A'].sum(),
        'C': peaksdb681['C'].sum(),
        'D': peaksdb681['D'].sum(),
        'E': peaksdb681['E'].sum(),
        'F': peaksdb681['F'].sum(),
        'G': peaksdb681['G'].sum(),
        'H': peaksdb681['H'].sum(),
        'I': peaksdb681['I'].sum(),
        'K': peaksdb681['K'].sum(),
        'L': peaksdb681['L'].sum(),
        'M': peaksdb681['M'].sum(),
        'N': peaksdb681['N'].sum(),
        'P': peaksdb681['P'].sum(),
        'Q': peaksdb681['Q'].sum(),
        'R': peaksdb681['R'].sum(),
        'S': peaksdb681['S'].sum(),
        'T': peaksdb681['T'].sum(),
        'V': peaksdb681['V'].sum(),
        'W': peaksdb681['W'].sum(),
        'Y': peaksdb681['Y'].sum(),
        'c-carb': peaksdb681['c-carb'].sum(),
        'm-oxid': peaksdb681['m-oxid'].sum(),
        'n-deam': peaksdb681['n-deam'].sum(),
        'q-deam': peaksdb681['q-deam'].sum(),
        'Total area': peaksdb681['Area'].sum(),
        'Total length': peaksdb681['stripped_length'].sum()
       }

totalpeaksdb681 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 
                                              'n-deam', 'q-deam', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb681['% C w/ carb'] = totalpeaksdb681['c-carb'] / totalpeaksdb681['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb681['% M w/ oxid'] = totalpeaksdb681['m-oxid'] / totalpeaksdb681['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb681['% N w/ deam'] = totalpeaksdb681['n-deam'] / totalpeaksdb681['N'] 

# calculate percentage of N's that are deamidated
totalpeaksdb681['% Q w/ deam'] = totalpeaksdb681['q-deam'] / totalpeaksdb681['Q'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb681['NAAF denom.'] = totalpeaksdb681['Total area'] / totalpeaksdb681['Total length']

# write modified dataframe to new txt file
totalpeaksdb681.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/681_NMCP_50_GF75_PDB/681_NMCP_50_GF75_PDB_totals.csv")

totalpeaksdb681.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb681 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 152144.612472

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb681['NAAF factor'] = (peaksdb681['NAAF_num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb681_NAAF = peaksdb681[['stripped_peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'n-deam', 'q-deam']].copy()

# multiply the NAAF20 factor NMCP the AAs to normalize its abundance NMCP peak area and peptide length

peaksdb681_NAAF['A-NAAF20'] = peaksdb681_NAAF['A'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['C-NAAF20'] = peaksdb681_NAAF['C'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['D-NAAF20'] = peaksdb681_NAAF['D'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['E-NAAF20'] = peaksdb681_NAAF['E'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['F-NAAF20'] = peaksdb681_NAAF['F'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['G-NAAF20'] = peaksdb681_NAAF['G'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['H-NAAF20'] = peaksdb681_NAAF['H'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['I-NAAF20'] = peaksdb681_NAAF['I'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['K-NAAF20'] = peaksdb681_NAAF['K'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['L-NAAF20'] = peaksdb681_NAAF['L'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['M-NAAF20'] = peaksdb681_NAAF['M'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['N-NAAF20'] = peaksdb681_NAAF['N'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['P-NAAF20'] = peaksdb681_NAAF['P'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['Q-NAAF20'] = peaksdb681_NAAF['Q'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['R-NAAF20'] = peaksdb681_NAAF['R'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['S-NAAF20'] = peaksdb681_NAAF['S'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['T-NAAF20'] = peaksdb681_NAAF['T'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['V-NAAF20'] = peaksdb681_NAAF['V'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['W-NAAF20'] = peaksdb681_NAAF['W'] * peaksdb681['NAAF factor']
peaksdb681_NAAF['Y-NAAF20'] = peaksdb681_NAAF['Y'] * peaksdb681['NAAF factor']

# multiply the NAAF20 factor NMCP the PTMs normalize its abundance NMCP peak area and peptide length

peaksdb681_NAAF['ccarb-NAAF20'] = peaksdb681_NAAF['c-carb'] * peaksdb681_NAAF['NAAF factor']
peaksdb681_NAAF['moxid-NAAF20'] = peaksdb681_NAAF['m-oxid'] * peaksdb681_NAAF['NAAF factor']
peaksdb681_NAAF['ndeam-NAAF20'] = peaksdb681_NAAF['n-deam'] * peaksdb681_NAAF['NAAF factor']
peaksdb681_NAAF['qdeam-NAAF20'] = peaksdb681_NAAF['q-deam'] * peaksdb681_NAAF['NAAF factor']


# write the dataframe to a new csv
peaksdb681_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/681_NMCP_50_GF75_PDB/681_NMCP_50_GF75_PDB_naaf.csv")

peaksdb681_NAAF.head()
```
