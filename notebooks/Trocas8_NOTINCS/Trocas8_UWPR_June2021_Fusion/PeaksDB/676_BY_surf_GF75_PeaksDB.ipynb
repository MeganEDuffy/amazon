### Manipulation of PeaskDB de novo-assisted database search results of Trocas 8 (low water, November 2019) non incubation samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Exported at <1.0% FDR
    
Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the TROCAS # (e.g., 675) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup676 = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/T8-June2021-UWPR-DN50_10lgP15/676_Trocas8_June2021_PEAKS_134/676_Trocas8_June2021_PEAKS_134_peptide.csv")

# remove redundant rows
peaksdb676 = pd.DataFrame.drop_duplicates(peaksdbdup676)

print(peaksdb676.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source_File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb676.columns = columns

#remmove # spec and accession columns because they mess parsing up

del peaksdb676['#Spec']
del peaksdb676['Accession']
del peaksdb676['PTM']
del peaksdb676['AScore']

mean_length = peaksdb676['Length'].mean()
print(mean_length)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup676))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb676))

#look at the dataframe
peaksdb676.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb676['A'] = peaksdb676['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb676['C'] = peaksdb676['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb676['D'] = peaksdb676['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb676['E'] = peaksdb676['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb676['F'] = peaksdb676['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb676['G'] = peaksdb676['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb676['H'] = peaksdb676['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb676 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb676['I'] = peaksdb676['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb676['K'] = peaksdb676['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb676['L'] = peaksdb676['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb676['M'] = peaksdb676['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb676['N'] = peaksdb676['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb676['P'] = peaksdb676['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb676['Q'] = peaksdb676['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb676['R'] = peaksdb676['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb676['S'] = peaksdb676['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb676['T'] = peaksdb676['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb676['V'] = peaksdb676['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb676['W'] = peaksdb676['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb676['Y'] = peaksdb676['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb676['c-carb'] = peaksdb676['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb676['m-oxid'] = peaksdb676['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb676['n-deam'] = peaksdb676['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a lamba function to enumerate the # of deamidated Q's in each peptide
peaksdb676['q-deam'] = peaksdb676['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb676['stripped_peptide'] = peaksdb676['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb676['stripped_length'] = peaksdb676['stripped_peptide'].apply(len)

peaksdb676['NAAF_num.'] = peaksdb676['Area'] / peaksdb676['stripped_length']

# total the number of modifications in sequence
peaksdb676['ptm-total'] = peaksdb676['c-carb'] + peaksdb676['m-oxid'] + peaksdb676['n-deam'] + peaksdb676['q-deam']

# turn all isoleucines into leucines
# this helps later in comparing Unipept peptides to PeaksDB and Comet ones
peaksdb676['stripped_IL']= peaksdb676['stripped_peptide'].str.replace('I','L')

# write modified dataframe to new txt file
peaksdb676.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/676_BY_surf_GF75_PDB/676_BY_surf_GF75_PDB.csv")

# check out the results
peaksdb676.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb676['A'].sum(),
        'C': peaksdb676['C'].sum(),
        'D': peaksdb676['D'].sum(),
        'E': peaksdb676['E'].sum(),
        'F': peaksdb676['F'].sum(),
        'G': peaksdb676['G'].sum(),
        'H': peaksdb676['H'].sum(),
        'I': peaksdb676['I'].sum(),
        'K': peaksdb676['K'].sum(),
        'L': peaksdb676['L'].sum(),
        'M': peaksdb676['M'].sum(),
        'N': peaksdb676['N'].sum(),
        'P': peaksdb676['P'].sum(),
        'Q': peaksdb676['Q'].sum(),
        'R': peaksdb676['R'].sum(),
        'S': peaksdb676['S'].sum(),
        'T': peaksdb676['T'].sum(),
        'V': peaksdb676['V'].sum(),
        'W': peaksdb676['W'].sum(),
        'Y': peaksdb676['Y'].sum(),
        'c-carb': peaksdb676['c-carb'].sum(),
        'm-oxid': peaksdb676['m-oxid'].sum(),
        'n-deam': peaksdb676['n-deam'].sum(),
        'q-deam': peaksdb676['q-deam'].sum(),
        'Total area': peaksdb676['Area'].sum(),
        'Total length': peaksdb676['stripped_length'].sum()
       }

totalpeaksdb676 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                              'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                                              'c-carb', 'm-oxid', 
                                              'n-deam', 'q-deam', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb676['% C w/ carb'] = totalpeaksdb676['c-carb'] / totalpeaksdb676['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb676['% M w/ oxid'] = totalpeaksdb676['m-oxid'] / totalpeaksdb676['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb676['% N w/ deam'] = totalpeaksdb676['n-deam'] / totalpeaksdb676['N'] 

# calculate percentage of N's that are deamidated
totalpeaksdb676['% Q w/ deam'] = totalpeaksdb676['q-deam'] / totalpeaksdb676['Q'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb676['NAAF denom.'] = totalpeaksdb676['Total area'] / totalpeaksdb676['Total length']

# write modified dataframe to new txt file
totalpeaksdb676.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/676_BY_surf_GF75_PDB/676_BY_surf_GF75_PDB_totals.csv")

totalpeaksdb676.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb676 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 143315.458915

# we'll use -10lgP > 20 as an approximate cutoff for the <1.0% FDR which is what we actually have
peaksdb676['NAAF factor'] = (peaksdb676['NAAF_num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb676_NAAF = peaksdb676[['stripped_peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', \
                              'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', \
                              'c-carb', 'm-oxid', 'n-deam', 'q-deam']].copy()

# multiply the NAAF20 factor by the AAs to normalize its abundance by peak area and peptide length

peaksdb676_NAAF['A-NAAF20'] = peaksdb676_NAAF['A'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['C-NAAF20'] = peaksdb676_NAAF['C'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['D-NAAF20'] = peaksdb676_NAAF['D'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['E-NAAF20'] = peaksdb676_NAAF['E'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['F-NAAF20'] = peaksdb676_NAAF['F'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['G-NAAF20'] = peaksdb676_NAAF['G'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['H-NAAF20'] = peaksdb676_NAAF['H'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['I-NAAF20'] = peaksdb676_NAAF['I'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['K-NAAF20'] = peaksdb676_NAAF['K'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['L-NAAF20'] = peaksdb676_NAAF['L'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['M-NAAF20'] = peaksdb676_NAAF['M'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['N-NAAF20'] = peaksdb676_NAAF['N'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['P-NAAF20'] = peaksdb676_NAAF['P'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['Q-NAAF20'] = peaksdb676_NAAF['Q'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['R-NAAF20'] = peaksdb676_NAAF['R'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['S-NAAF20'] = peaksdb676_NAAF['S'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['T-NAAF20'] = peaksdb676_NAAF['T'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['V-NAAF20'] = peaksdb676_NAAF['V'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['W-NAAF20'] = peaksdb676_NAAF['W'] * peaksdb676['NAAF factor']
peaksdb676_NAAF['Y-NAAF20'] = peaksdb676_NAAF['Y'] * peaksdb676['NAAF factor']

# multiply the NAAF20 factor by the PTMs normalize its abundance by peak area and peptide length

peaksdb676_NAAF['ccarb-NAAF20'] = peaksdb676_NAAF['c-carb'] * peaksdb676_NAAF['NAAF factor']
peaksdb676_NAAF['moxid-NAAF20'] = peaksdb676_NAAF['m-oxid'] * peaksdb676_NAAF['NAAF factor']
peaksdb676_NAAF['ndeam-NAAF20'] = peaksdb676_NAAF['n-deam'] * peaksdb676_NAAF['NAAF factor']
peaksdb676_NAAF['qdeam-NAAF20'] = peaksdb676_NAAF['q-deam'] * peaksdb676_NAAF['NAAF factor']


# write the dataframe to a new csv
peaksdb676_NAAF.to_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/PeaksDB/676_BY_surf_GF75_PDB/676_BY_surf_GF75_PDB_naaf.csv")

peaksdb676_NAAF.head()
```
