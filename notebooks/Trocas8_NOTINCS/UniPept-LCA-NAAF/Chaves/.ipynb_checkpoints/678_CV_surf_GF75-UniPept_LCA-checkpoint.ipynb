```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
cd /home/millieginty/Documents/git-repos/amazon/analyses/T8-notincs/unipept/
```

```python
# read the data into pandas dataframes
CV_surf_GF75_678a = pd.read_csv("lca/cleaned/678_CV_surf_GF75_lca.csv")

# delete any Metazoa hits because they're trypsin
CV_surf_GF75_678b = CV_surf_GF75_678a[CV_surf_GF75_678a.kingdom != 'Metazoa']

# keep only entries to the phylum level
CV_surf_GF75_678 = CV_surf_GF75_678b[CV_surf_GF75_678b['phylum'].notnull()]

# How many peptides to the phylum level?
print('# of phylum peptides = ', len(CV_surf_GF75_678))

CV_surf_GF75_678.head(6)
```

### Now I want to read in the file containing the stripped peptides with NAAF values
####  - NAAF stands for 'noramlized area abunace factor'

### I want to join the dataframes if they share an index (stripped peptide with equated leucine and isoleucines)
#### - That means I'll reindex the processed peptide file

```python
CV_surf_GF75_678_NAAFa = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/678_CV_surf_GF75_DN50_ILnaafs.csv")
CV_surf_GF75_678_NAAFb = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/678_CV_surf_GF75_PDB_ILnaafs.csv")
CV_surf_GF75_678_NAAFc = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/686_CV_surf_GF75_DN50_ILnaafs.csv")
CV_surf_GF75_678_NAAFd = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/686_CV_surf_GF75_PDB_ILnaafs.csv")


frames = [CV_surf_GF75_678_NAAFa, CV_surf_GF75_678_NAAFb, CV_surf_GF75_678_NAAFc, CV_surf_GF75_678_NAAFd]

CV_surf_GF75_678_NAAF = pd.concat(frames, sort=False)

CV_surf_GF75_678_NAAF.set_index('stripped_peptide')
CV_surf_GF75_678_NAAF = CV_surf_GF75_678_NAAF.loc[:, ~CV_surf_GF75_678_NAAF.columns.str.contains('^Unnamed')]

CV_surf_GF75_678_NAAF.rename(columns = {'stripped_peptide':'peptide'}, inplace = True)

print('# of total peptides = ', len(CV_surf_GF75_678_NAAF))

print('column names:', CV_surf_GF75_678_NAAF.columns)

CV_surf_GF75_678_NAAF.head()
```

```python
CV_surf_GF75_678_NAAF['peptide'] = CV_surf_GF75_678_NAAF['peptide'].astype(str)
CV_surf_GF75_678['peptide'] = CV_surf_GF75_678['peptide'].astype(str)
```

```python
# get the intersection of the phylum peptides and corresponding peptides w/ NAAFs

over = np.intersect1d(CV_surf_GF75_678["peptide"], CV_surf_GF75_678_NAAF["peptide"])

print(over)
```

```python
# join the dataframes if the peptide values are the same using 'join'
# since a couple are de novo only (more for bacteria), we won't have all the UniPept peptides overlap 

CV_surf_GF75_678.set_index('peptide', inplace=True)
CV_surf_GF75_678_NAAF.set_index('peptide', inplace=True)

CV_surf_GF75_678_Phy = CV_surf_GF75_678.join(CV_surf_GF75_678_NAAF, how='left', rsuffix='_other')


print('# of total phylum-level peptides = ', len(CV_surf_GF75_678_Phy))

CV_surf_GF75_678_Phy.head()
```

```python
# write to a csv

CV_surf_GF75_678_Phy.to_csv("lca/NAAF/CV_surf_GF75_678_Phy_naaf.csv")

!ls lca/NAAF/
```
