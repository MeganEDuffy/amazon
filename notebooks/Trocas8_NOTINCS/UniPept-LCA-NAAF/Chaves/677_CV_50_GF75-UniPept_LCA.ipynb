```python
# read the data into pandas dataframes
CV_50_GF75_677a = pd.read_csv("lca/cleaned/677_CV_50_GF75_lca.csv")

# delete any Metazoa hits because they're trypsin
CV_50_GF75_677b = CV_50_GF75_677a[CV_50_GF75_677a.kingdom != 'Metazoa']

# keep only entries to the phylum level
CV_50_GF75_677 = CV_50_GF75_677b[CV_50_GF75_677b['phylum'].notnull()]

# How many peptides to the phylum level?
print('# of phylum peptides = ', len(CV_50_GF75_677))

CV_50_GF75_677.head(6)
```

### Now I want to read in the file containing the stripped peptides with NAAF values
####  - NAAF stands for 'noramlized area abunace factor'

### I want to join the dataframes if they share an index (stripped peptide with equated leucine and isoleucines)
#### - That means I'll reindex the processed peptide file

```python
CV_50_GF75_677_NAAFa = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/677_CV_50_GF75_DN50_ILnaafs.csv")
CV_50_GF75_677_NAAFb = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/677_CV_50_GF75_PDB_ILnaafs.csv")
CV_50_GF75_677_NAAFc = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/685_CV_50_GF75_DN50_ILnaafs.csv")
CV_50_GF75_677_NAAFd = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/685_CV_50_GF75_PDB_ILnaafs.csv")


frames = [CV_50_GF75_677_NAAFa, CV_50_GF75_677_NAAFb, CV_50_GF75_677_NAAFc, CV_50_GF75_677_NAAFd]

CV_50_GF75_677_NAAF = pd.concat(frames, sort=False)

CV_50_GF75_677_NAAF.set_index('stripped_peptide')
CV_50_GF75_677_NAAF = CV_50_GF75_677_NAAF.loc[:, ~CV_50_GF75_677_NAAF.columns.str.contains('^Unnamed')]

CV_50_GF75_677_NAAF.rename(columns = {'stripped_peptide':'peptide'}, inplace = True)

print('# of total peptides = ', len(CV_50_GF75_677_NAAF))

print('column names:', CV_50_GF75_677_NAAF.columns)

CV_50_GF75_677_NAAF.head()
```

```python
CV_50_GF75_677_NAAF['peptide'] = CV_50_GF75_677_NAAF['peptide'].astype(str)
CV_50_GF75_677['peptide'] = CV_50_GF75_677['peptide'].astype(str)
```

```python
# get the intersection of the phylum peptides and corresponding peptides w/ NAAFs

over = np.intersect1d(CV_50_GF75_677["peptide"], CV_50_GF75_677_NAAF["peptide"])

print(over)
```

```python
# join the dataframes if the peptide values are the same using 'join'
# since a couple are de novo only (more for bacteria), we won't have all the UniPept peptides overlap 

CV_50_GF75_677.set_index('peptide', inplace=True)
CV_50_GF75_677_NAAF.set_index('peptide', inplace=True)

CV_50_GF75_677_Phy = CV_50_GF75_677.join(CV_50_GF75_677_NAAF, how='left', rsuffix='_other')


print('# of total phylum-level peptides = ', len(CV_50_GF75_677_Phy))

CV_50_GF75_677_Phy.head()
```

```python
# write to a csv

CV_50_GF75_677_Phy.to_csv("lca/NAAF/CV_50_GF75_677_Phy_naaf.csv")

!ls lca/NAAF/
```
