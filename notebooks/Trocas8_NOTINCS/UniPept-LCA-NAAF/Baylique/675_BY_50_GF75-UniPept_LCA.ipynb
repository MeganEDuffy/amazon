```python
# read the data into pandas dataframes
BY_50_GF75_675a = pd.read_csv("lca/cleaned/675_BY_50_GF75_lca.csv")

# delete any Metazoa hits because they're trypsin
BY_50_GF75_675b = BY_50_GF75_675a[BY_50_GF75_675a.kingdom != 'Metazoa']

# keep only entries to the phylum level
BY_50_GF75_675 = BY_50_GF75_675b[BY_50_GF75_675b['phylum'].notnull()]

# How many peptides to the phylum level?
print('# of phylum peptides = ', len(BY_50_GF75_675))

BY_50_GF75_675.head(6)
```

### Now I want to read in the file containing the stripped peptides with NAAF values
####  - NAAF stands for 'noramlized area abunace factor'

### I want to join the dataframes if they share an index (stripped peptide with equated leucine and isoleucines)
#### - That means I'll reindex the processed peptide file

```python
BY_50_GF75_675_NAAFa = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/675_BY_50_GF75_DN50_ILnaafs.csv")
BY_50_GF75_675_NAAFb = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/675_BY_50_GF75_PDB_ILnaafs.csv")
BY_50_GF75_675_NAAFc = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/683_BY_50_GF75_DN50_ILnaafs.csv")
BY_50_GF75_675_NAAFd = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/683_BY_50_GF75_PDB_ILnaafs.csv")


frames = [BY_50_GF75_675_NAAFa, BY_50_GF75_675_NAAFb, BY_50_GF75_675_NAAFc, BY_50_GF75_675_NAAFd]

BY_50_GF75_675_NAAF = pd.concat(frames, sort=False)

BY_50_GF75_675_NAAF.set_index('stripped_peptide')
BY_50_GF75_675_NAAF = BY_50_GF75_675_NAAF.loc[:, ~BY_50_GF75_675_NAAF.columns.str.contains('^Unnamed')]

BY_50_GF75_675_NAAF.rename(columns = {'stripped_peptide':'peptide'}, inplace = True)

print('# of total peptides = ', len(BY_50_GF75_675_NAAF))

print('column names:', BY_50_GF75_675_NAAF.columns)

BY_50_GF75_675_NAAF.head()
```

```python
BY_50_GF75_675_NAAF['peptide'] = BY_50_GF75_675_NAAF['peptide'].astype(str)
BY_50_GF75_675['peptide'] = BY_50_GF75_675['peptide'].astype(str)
```

```python
# get the intersection of the phylum peptides and corresponding peptides w/ NAAFs

over = np.intersect1d(BY_50_GF75_675["peptide"], BY_50_GF75_675_NAAF["peptide"])

print(over)
```

```python
# join the dataframes if the peptide values are the same using 'join'
# since a couple are de novo only (more for bacteria), we won't have all the UniPept peptides overlap 

BY_50_GF75_675.set_index('peptide', inplace=True)
BY_50_GF75_675_NAAF.set_index('peptide', inplace=True)

BY_50_GF75_675_Phy = BY_50_GF75_675.join(BY_50_GF75_675_NAAF, how='left', rsuffix='_other')


print('# of total phylum-level peptides = ', len(BY_50_GF75_675_Phy))

BY_50_GF75_675_Phy.head()
```

```python
# write to a csv

BY_50_GF75_675_Phy.to_csv("lca/NAAF/BY_50_GF75_675_Phy_naaf.csv")

!ls lca/NAAF/
```
