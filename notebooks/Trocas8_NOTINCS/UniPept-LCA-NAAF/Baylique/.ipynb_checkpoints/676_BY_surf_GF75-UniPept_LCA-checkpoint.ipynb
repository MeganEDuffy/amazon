```python
# read the data into pandas dataframes
BY_surf_GF75_676a = pd.read_csv("lca/cleaned/676_BY_surf_GF75_lca.csv")

# delete any Metazoa hits because they're trypsin
BY_surf_GF75_676b = BY_surf_GF75_676a[BY_surf_GF75_676a.kingdom != 'Metazoa']

# keep only entries to the phylum level
BY_surf_GF75_676 = BY_surf_GF75_676b[BY_surf_GF75_676b['phylum'].notnull()]

# How many peptides to the phylum level?
print('# of phylum peptides = ', len(BY_surf_GF75_676))

BY_surf_GF75_676.head(6)
```

### Now I want to read in the file containing the stripped peptides with NAAF values
####  - NAAF stands for 'noramlized area abunace factor'

### I want to join the dataframes if they share an index (stripped peptide with equated leucine and isoleucines)
#### - That means I'll reindex the processed peptide file

```python
BY_surf_GF75_676_NAAFa = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/676_BY_surf_GF75_DN50_ILnaafs.csv")
BY_surf_GF75_676_NAAFb = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/676_BY_surf_GF75_PDB_ILnaafs.csv")
BY_surf_GF75_676_NAAFc = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/684_BY_surf_GF75_DN50_ILnaafs.csv")
BY_surf_GF75_676_NAAFd = pd.read_csv("/home/millieginty/Documents/git-repos/amazon/data/Trocas8-notincs/processed/I-L_NAAFs/684_BY_surf_GF75_PDB_ILnaafs.csv")


frames = [BY_surf_GF75_676_NAAFa, BY_surf_GF75_676_NAAFb, BY_surf_GF75_676_NAAFc, BY_surf_GF75_676_NAAFd]

BY_surf_GF75_676_NAAF = pd.concat(frames, sort=False)

BY_surf_GF75_676_NAAF.set_index('stripped_peptide')
BY_surf_GF75_676_NAAF = BY_surf_GF75_676_NAAF.loc[:, ~BY_surf_GF75_676_NAAF.columns.str.contains('^Unnamed')]

BY_surf_GF75_676_NAAF.rename(columns = {'stripped_peptide':'peptide'}, inplace = True)

print('# of total peptides = ', len(BY_surf_GF75_676_NAAF))

print('column names:', BY_surf_GF75_676_NAAF.columns)

BY_surf_GF75_676_NAAF.head()
```

```python
BY_surf_GF75_676_NAAF['peptide'] = BY_surf_GF75_676_NAAF['peptide'].astype(str)
BY_surf_GF75_676['peptide'] = BY_surf_GF75_676['peptide'].astype(str)
```

```python
# get the intersection of the phylum peptides and corresponding peptides w/ NAAFs

over = np.intersect1d(BY_surf_GF75_676["peptide"], BY_surf_GF75_676_NAAF["peptide"])

print(over)
```

```python
# join the dataframes if the peptide values are the same using 'join'
# since a couple are de novo only (more for bacteria), we won't have all the UniPept peptides overlap 

BY_surf_GF75_676.set_index('peptide', inplace=True)
BY_surf_GF75_676_NAAF.set_index('peptide', inplace=True)

BY_surf_GF75_676_Phy = BY_surf_GF75_676.join(BY_surf_GF75_676_NAAF, how='left', rsuffix='_other')


print('# of total phylum-level peptides = ', len(BY_surf_GF75_676_Phy))

BY_surf_GF75_676_Phy.head()
```

```python
# write to a csv

BY_surf_GF75_676_Phy.to_csv("lca/NAAF/BY_surf_GF75_676_Phy_naaf.csv")

!ls lca/NAAF/
```
